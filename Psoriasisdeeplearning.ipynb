```python
import cv2
import os
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
path=os.listdir('/content/drive/MyDrive/Psoriasis123')
classes={'YES':1,'NO':0}
```

```python
X=[]
Y=[]
for i in classes:
  path1='/content/drive/MyDrive/Psoriasis123/'+i
  for j in os.listdir(path1):
    img=cv2.imread(path1+'/'+j,0)
    img=cv2.resize(img,(128,128))
    img= cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    X.append(img)
    Y.append(classes[i])
```

```python
X=np.array(X)
Y=np.array(Y)
```

ResNet 32

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Activation, Add, AveragePooling2D, Flatten, Dense
from tensorflow.keras.models import Model

# Define the Residual Block
def residual_block(x, filters, stride=1):
    shortcut = x
    x = Conv2D(filters, (3, 3), strides=stride, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Second convolution layer
    x = Conv2D(filters, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)

    # If the input shape and output shape don't match, use a 1x1 convolution for the shortcut
    if stride != 1 or shortcut.shape[-1] != filters:
        shortcut = Conv2D(filters, (1, 1), strides=stride)(shortcut)

    # Add the shortcut to the main path
    x = Add()([x, shortcut])
    x = Activation('relu')(x)
    return x

# Define the ResNet-32 model
def resnet32(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    x = Conv2D(32, (3, 3), padding='same')(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Stack of residual blocks
    num_blocks_list = [5, 5, 5]
    for num_blocks in num_blocks_list:
        for _ in range(num_blocks):
            x = residual_block(x, 32)

    # Global Average Pooling
    x = AveragePooling2D(pool_size=(8, 8))(x)

    # Fully connected layer
    x = Flatten()(x)
    x = Dense(num_classes, activation='softmax')(x)

    # Create the model
    model = Model(inputs, x)
    return model

# Create the ResNet-32 model
model = resnet32(input_shape=(128, 128, 3), num_classes=2)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()

# Train the model using X and Y
model.fit(X, Y, epochs=10, batch_size=32, validation_split=0.2)
```
